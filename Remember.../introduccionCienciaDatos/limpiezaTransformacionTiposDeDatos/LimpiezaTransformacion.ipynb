```python
## tratamiento de valores nulos


# elimina filas que contienen al menos un valor nulo
df_sin_nulos = salaries.dropna() 


# elimina columnas que contienen al menos un valor nulo
df_sin_nulos_col = salaries.dropna(axis=1)
```

```python
df_sin_nulos.shape
```

```python
# transformación de datos
# 2da opc reemplazar los NaNs por la media

# parsing ("Parseo") de la columna Salary

df_sin_nulos['salary']
```

```python
## este codigo toma una columna 'salary' que contiene valores de salario en formato cadena string
## elimina cualquier caracter no numerico, extrae los digitos y convierte la columna resultante en valores numéricos de tipo float
## el resultado se almacena en una nueva columna llama 'salary_numeric'
## esto es util cuando la columna 'salary' inicial contiene caracteres no numéricos y se desea trabajar con los salarios como valores numéricos


df_sin_nulos['salary_numeric'] = df_sin_nulos['salary'].replace('[^\d.]', '', regex=True).str.extract('(\d+)').astype(float)
```

```python
df_sin_nulos['salary_numeric']
```

```python
df_sin_nulos.describe()
```

```python
# crear un boxplot
plt.boxplot(df_sin_nulos['salary_numeric'])

# añadir etiquetas y titulo
plt.xlabel('Salarios')
plt.ylabel('Valores en USD')
plt.title('BoxPlot')

plt.show()

# distribución de valores de está columna
```

```python
# crear un histograma 
plt.hist(df_sin_nulos['salary_numeric'], bins=30, color='blue', alpha=0.7)

# añadir etiquetas y titulo
plt.xlabel('valores')
plt.ylabel('frecuencia')
plt.title('Histograma de datos aleatorios')

plt.show()
```

```python
## como hay salarios x/hora /año

# eliminamos las filas que contengan un salario mayor o igual a 20000

df_sin_nulos = df_sin_nulos[df_sin_nulos['salary_numeric'] >= 20000]
```

```python
# crear un histograma 
plt.hist(df_sin_nulos['salary_numeric'])

# añadir etiquetas y titulo
plt.xlabel('valores')
plt.ylabel('frecuencia')
plt.title('Histograma de datos aleatorios')

plt.show()
```

```python
# metodo de reemplazo de valores nulos con la media de la columna 
df_con_relleno = df_sin_nulos.fillna(df_sin_nulos['salary_numeric'].mean())
```

```python
## eliminar la columna salary deprecada

df_con_relleno.drop('salary', axis=1, inplace=True)
```

```python
df_con_relleno
```

```python
# AGRUPAR Y AGREGAR DATOS

# supongamos que tienes un dataframe llamado df con col 'title' y 'salary'ArithmeticError

df_grouped = df_con_relleno.groupby('title')['salary_numeric'].agg(['mean', 'median', 'count']).reset_index()

df_grouped.columns = ['title', 'salary', 'salary_median', 'job_count']

df_grouped
```

```python
df_grouped['salary'] = df_grouped['salary'].round(0).astype(int)
df_grouped['salary_median'] = df_grouped['salary_median'].round(0).astype(int)
```

```python
df_grouped
```

```python
# exportar csv

pathAndFileName = "export-linkedin-jobs-usa"
df_grouped.to_csv(pathAndFileName)
```
