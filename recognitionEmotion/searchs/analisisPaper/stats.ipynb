```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
from matplotlib_venn import venn3

set1 = {"Hardware"}
set2 = {"Biomedical applications", "Drug therapy", "Disease classification"}
set3 = {"Adults", "Elderly", "Animals"}

venn3([set1, set2, set3], ('Por tecnicismos', 'Por léxico Médico', 'Por edad y organismo'))

plt.title("Diagrama de Venn de Términos de Exclusión")
plt.show()
```

# Términos de búsqueda combinados (conjuntos)

```python
venn = venn3(subsets=(1, 1, 1, 1, 1, 1, 1), 
              set_labels=('Deep Learning', 'Emotion Detection', 'Children'))

for i in venn.subset_labels:
    if i is not None:
        i.set_text('') 

plt.title("Diagrama de Venn: Términos de Inclusión")
plt.show()
```

```python
# Neural Networks AND Emotion classification AND Adolescents
venn = venn3(subsets=(1, 1, 1, 1, 1, 1, 1), 
              set_labels=('Neuronal Networks', 'Emotion classification', 'Adolescents'))

for i in venn.subset_labels:
    if i is not None:
        i.set_text('') 

plt.title("Diagrama de Venn: Términos de Inclusión")
plt.show()
```

```python
# Sentiment analysis AND Neural Networks AND Children OR Adolescents
venn = venn3(subsets=(1, 1, 1, 1, 1, 1, 1), 
              set_labels=('Sentiment analysis', 'Neural Networks', 'Children'))

for i in venn.subset_labels:
    if i is not None:
        i.set_text('') 

plt.title("Diagrama de Venn: Términos de Inclusión")
plt.show()
```

```python
plt.figure(figsize=(10, 6))

inclusion1 = plt.Circle((0.3, 0.5), 0.2, color='lightblue', label='Deep Learning\nAND Emotion detection\nAND Children')
inclusion2 = plt.Circle((0.5, 0.5), 0.2, color='lightgreen', label='Neural Networks\nAND Emotion classification\nAND Adolescents', alpha=0.5)
inclusion3 = plt.Circle((0.4, 0.3), 0.2, color='lightcoral', label='Deep Learning\nAND Affective computing\nAND Pediatric', alpha=0.5)

exclusion = plt.Circle((0.8, 0.5), 0.08, color='orange', label='Adults (Excluded)', alpha=0.5)

ax = plt.gca()
ax.add_artist(inclusion1)
ax.add_artist(inclusion2)
ax.add_artist(inclusion3)
ax.add_artist(exclusion)

plt.xlim(0, 1)
plt.ylim(0, 1)

plt.title("Diagrama de Conjuntos: Términos de Inclusión y Exclusión")
plt.legend(loc='upper left')
plt.axis('off')  
plt.show()
```
