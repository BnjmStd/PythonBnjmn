```python
# Papers por año 
papersPorAno = dataFrame['Año de Publicacion'].value_counts()
print(papersPorAno)
```

```python
# Ordenar los papers por año
papersPorAno_sorted = papersPorAno.sort_index()

# Crear una función para mostrar tanto el número de papers como el porcentaje dentro del gráfico
def func(pct, allvals):
    absolute = int(pct/100. * sum(allvals))
    return f'{absolute} ({pct:.1f}%)'

# Crear el gráfico de torta
plt.figure(figsize=(8, 6))
plt.pie(papersPorAno_sorted, labels=papersPorAno_sorted.index, 
        autopct=lambda pct: func(pct, papersPorAno_sorted), startangle=140)
plt.title('Distribución de Papers por Año')
plt.axis('equal')  # Asegura que el gráfico sea un círculo
plt.show()
```
